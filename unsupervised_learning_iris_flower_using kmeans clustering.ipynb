import the libaries needed for this excercise 

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

load the iris dataset from sklearn datasets

```
iris_dataset = load_iris()
```

lets see what columns this dataset contains

```
print(iris_dataset.keys())
```

lets convert this dataset into a pandas dataframe

```
df = pd.DataFrame(data=np.c_[iris_dataset["data"]],columns=iris_dataset["feature_names"])
df["targets"] = iris_dataset["target"]
```

```
df.head()
```

```
X = df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]]
Y=df["targets"]
```

we will do a elbow graph to determine what is the best number of clusters to use, to read the elbow chart look at the second bend in the graph, the y axis is the Sum of squared distances of samples to their closest cluster center or 
WCSS(Within Cluster Sum of Squares
) and the x axis is th number of clusters 

```
elbow = []
for i in range(1,11):
  kmean = KMeans(n_clusters=i).fit(df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]]).inertia_
  elbow.append(kmean)
```

as we can see from the elbow graph the second bend isthw optiaml number of clusters in this case it is 3 clusters 

```
plt.plot(range(1,11),elbow)
plt.title("Elbow graph")
plt.xlabel("no of clusters")
plt.ylabel("inertia")
```

```
model = KMeans(n_clusters=3)
model_pred = model.fit_predict(X)
```
